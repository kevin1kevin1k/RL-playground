```python
import gym
```

```python
env = gym.make('CartPole-v0')
# env = gym.make('MountainCar-v0')

for i_ep in range(3):
    env.reset()
    done = False
    while not done:
        env.render()
        action = env.action_space.sample()
        observation, reward, done, info = env.step(action)
        print(observation)
    env.close()
```
